```python
import IPython # DataFrame visualization
import json # import support files
import pandas # DataFrame functions and methods
```

```python
with open("terms.json", "r") as translation_file:
    term = json.load(translation_file)
```

```python
class Event:
    init_data = { # parameter of Event
        "Time_Stamp": "未知",
        "Institution": "其他",
        "Title": "無標題",
        "Catagory": "其他",
        "Debit": 0,
        "Credit": 0,
        "Remark": "無"
    }

    def __init__(self, **entry): # Event initialization
        self.__dict__.update(self.init_data)
        for key, value in entry.items(): # support Event info imported from initialization
            if key in self.__dict__:
                self.__dict__[key] = value

    def get_TimeStamp(self):
        try:
            data = input(f"{term['Time_Stamp']}為").strip()
            if data != "": self.Time_Stamp = data
        except KeyboardInterrupt:
            print("輸入中斷存檔失敗")
        except:
            print("發生未知錯誤")

    def get_Institution(self):
        try:
            data = input(f"{term['Institution']}為").strip()
            if data != "": self.Institution = data
        except KeyboardInterrupt:
            print("輸入中斷存檔失敗")
        except:
            print("發生未知錯誤")

    def get_Title(self):
        try:
            data = input(f"{term['Title']}為").strip()
            if data != "": self.Title = data
        except KeyboardInterrupt:
            print("輸入中斷存檔失敗")
        except:
            print("發生未知錯誤")

    def get_Catagory(self):
        try:
            data = input(f"{term['Catagory']}為").strip()
            if data != "": self.Catagory = data
        except KeyboardInterrupt:
            print("輸入中斷存檔失敗")
        except:
            print("發生未知錯誤")

    def get_Debit(self):
        try:
            data = input(f"{term['Debit']}為").strip()
            if data != "": self.Debit = int(data)
        except KeyboardInterrupt:
            print("輸入中斷存檔失敗")
        except:
            print("發生未知錯誤")

    def get_Credit(self):
        try:
            data = input(f"{term['Credit']}為").strip()
            if data != "": self.Credit = (data)
        except KeyboardInterrupt:
            print("輸入中斷存檔失敗")
        except:
            print("發生未知錯誤")

    def get_Remark(self):
        try:
            data = input(f"{term['Remark']}為").strip()
            if data != "": self.Remark = data
        except KeyboardInterrupt:
            print("輸入中斷存檔失敗")
        except:
            print("發生未知錯誤")

    def create(self): # manual input Event info and make it DataFrame
        self.get_TimeStamp()
        self.get_Institution()
        self.get_Title()
        self.get_Catagory()
        self.get_Debit()
        self.get_Credit()
        self.get_Remark()
        data = pandas.DataFrame([self.__dict__]) # turn the Event attributes into a single row of DataFrame
        IPython.display.display(data) # show the result to the user
        return data
```
