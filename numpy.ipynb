```python
import numpy as np
```

```python
data1 = np.array([1, 2, 3, 4, 5])
data2 = np.array([5, 4, 3, 2, 1])
data3 = data1 * data2 #[5 8 9 8 5]
print(sum(data3))
print(np.dot(data1, data2))
```

```python
A = np.random.randn(50)
print(A)
print("----------")
A = A * 2 + 5
print(A)
print("----------")
print(A.shape)
A.shape = (10, 5)
print(A)
print("----------")
B = A.reshape(5, 10)
print(B)
```

```python
C = np.zeros(6)
print(C)
print("---------")
C = np.zeros((3, 8))
print(C)
print("---------")
C = np.ones(6)
print(C)
print("---------")
C = np.ones((3, 8))
print(C)
print("---------")
C = np.eye(5)
print(C)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot(np.array([np.random.randn() for x in range(100)]), "bx-", markersize=6)
```

```python
curve = np.linspace(-2*np.pi, 2*np.pi, 100) #遞增陣列(開始，結束，數量)
plt.plot(curve, curve/4, color="red")
plt.plot(curve, np.sin(curve), color="green")
plt.plot(curve, np.cos(curve), color="blue")
```

```python
shuffle = np.array([4, -2, 5, -3, 100, -23, 17, -66]) #Array 快篩法
print(shuffle>0)
print(shuffle<0)
print(shuffle[shuffle>0])
```

```python
x = np.linspace(-5, 5, 2000)
y = np.sinc(x)
plt.plot(x, y, color="gray")
plt.plot(x[y>0], y[y>0]-0.3, "o", markersize=2, color="black")
plt.plot(x[x>0], y[x>0]+0.3, "o", markersize=2, color="violet")
```
