```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact
from ipywidgets import interact_manual
```

```python
class functionSet:
    def __init__(self):
        return None

    def Int_Output(self, x:int):
        return x

    def Str_Output(self, x:str):
        return x

    def List_Output(self, x:list):
        return x

    def Dict_Mapping(self, x:dict):
        return x

    def Move(self, x:int):
        print(" "*x + "o")

sample = functionSet()
```

```python
interact(sample.Int_Output, x=(0, 10))
```

```python
interact(sample.Str_Output, x="Sal")
```

```python
List = ["Salmon", "Tuna", "Martin", "Eel"]
interact(sample.List_Output, x=List)
```

```python
Dict = {"Salmon":500, "Tuna":320, "Martin":450, "Eel":200}
interact(sample.Dict_Mapping, x=Dict)
```

```python
interact(sample.Move, x=(0, 10))
```

```python
def Tsin(n:int):
    x = np.linspace(0, 5*np.pi, 1000)
    y = np.sin(n*x)
    plt.plot(x, y)
    plt.show()

interact(Tsin, n=5)
```

```python
interact_manual(Tsin, n=(0., 20.))
```

```python
def BMIcal(height:int, weight:int):
    bmi = (weight/(height/100)**2)
    print(f"Your bmi is {round(bmi, 2)}")

interact_manual(BMIcal, height=168.2, weight=56.)
```
