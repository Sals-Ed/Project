```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 50)
y = 2 * x + 1.2 * np.random.randn(50)
plt.scatter(x, y)
plt.plot(x, 2 * x, "r")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x = np.linspace(1, 5, 100)
y = np.linspace(1, 5, 100)
y = np.array([element + np.random.rand() for element in y])

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=66)
x_train.shape = (80, 1); x_test.shape = (20, 1)
y_train.shape = (80, 1); y_test.shape = (20, 1)
```

```python
regr = LinearRegression()
regr.fit(x_train, y_train)
plt.scatter(x_train, y_train)
plt.plot(x_train, regr.predict(x_train), "r", linestyle="dashed")
```

```python
plt.scatter(x_test, y_test)
plt.plot(x_test, regr.predict(x_test), "r", linestyle="dashed")
```

```python
from sklearn.svm import SVC
```

```python
train = np.array([[1, 1], [2, 0.5], [3, 1], [4, 2], [5, 1]])
category = np.array([1, 1, 1, 2, 2])

plt.scatter(train[:,0], train[:,1], c=category)
```

```python
clf = SVC()
clf.fit(train, category)
```

```python
formX, formY = np.meshgrid(np.linspace(0, 6, 30), np.linspace(-2, 4, 30))
formX = formX.ravel(); formY = formY.ravel()

plt.scatter(formX, formY, c=clf.predict(list(zip(formX, formY))))
```
