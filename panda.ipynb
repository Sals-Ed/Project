```python
column_name = ["Rank", "Old"]
row_name = ["Aquaman", "Yappy", "Gorzilla", "Zenbo", "Sowut", "Suhf", "Gummy B", "Savage.M", "AMILI", "Macdella"]
data = [[2, 16], [3, 18], [1, 20], [4, 8], [5, 32]]

Dashboard1 = pd.DataFrame(data, columns=column_name, index=row_name[0:5])
print(Dashboard1)
print("----------")

Dashboard2 = Dashboard1.copy()
Dashboard2["Rank"] += 5
Dashboard2["Old"] += 10
Dashboard2.index = row_name[5:10]
print(Dashboard2)
print("----------")

TotalRank = pd.concat([Dashboard1, Dashboard2])
TotalRank.sort_values(by=["Rank"], inplace=True)
TotalRank.to_csv("Rap.csv")

TotalRank.loc["Zenbo":"Suhf", "Rank":"Old"] # 2D Slice
```
