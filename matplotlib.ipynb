```python
#%matplotlib inline
import matplotlib.pyplot as plt
```

`matplotlib.pyplot.plot`

> `func(*args, scalex=True, scaley=True, data=None, **kwargs)`

* *args

    以 ([x] : list, y : list, [fmt] : string) 為一條線的單位

    [fmt] = "ob, og, or", "bo, go, ro", "bx, gx, rx", "b+, g+, r+", "^k:"

* scale : bool

    scalex, scaley，分別代表是否將**xy軸**參考**資料集**標齊

* data : dict

    可將dict中資料映射給 plot 繪圖

* **kwargs

    常見參數如下

    color : string / tuple( :int / 255, :int / 255, :int / 255)

    marker : string = "o" , "x" , "+"

    markersize : int

    linestyle : int

    linewidth : int

```python
plt.plot([1, 3, 4, 5, 6], [5, 3, 1, 6, 7], "or", [1, 2, 3, 4, 5], [4, 3, 1, 6, 7], "b+", linestyle="--")
```

```python
dataset = {"XSet1":[1, 2, 3, 4, 5], "YSet1":[3, 2, 1, 4, 5], 'XSet2':[1, 2, 3, 4, 5], "YSet2":[2, 3, 1, 6, 7]}
plt.plot("XSet1", "YSet1", color="purple", data=dataset)
plt.plot("XSet2", "YSet2", color="violet", data=dataset)
```
